```python
import pandas as pd
import glob
import os

# setting the path for joining multiple files
files = os.path.join("C:\\Users\\MAYANK\\Downloads\\", "customer_data_*.csv")

# list of merged files returned
files = glob.glob(files)

print("Resultant CSV after joining all CSV files at a particular location...");

# joining files with concat and read_csv and remove duplicates from file one
df = pd.concat(map(pd.read_csv, files), ignore_index=True).drop_duplicates(subset='first_name', keep="last").reset_index(drop=True)
df.head()
```

```python
# Sorting values and resetting index

df = df.sort_values(by='first_name', ascending=True).reset_index(drop=True)
df.head()
```

```python
# converting dataframe to json objects

import json
json = df.to_json()
print(json)
```

```python
#coverting to split formatted json

result = df.to_json(orient="split")
print(result)
```
